# Unit 4 Data Cleaning Lecturecast Exercise
I used UNICEF's Child Labour datasets to practise data cleaning using Python. This included replacing acronym headers with more readable equivalents (found in the mn_headers.csv file). I also experimented using Python's zip method to do this. 

```python
# Replacing headers
from csv import DictReader
data_rdr=DictReader(open("mn.csv"))
header_rdr=DictReader(open("mn_headers.csv"))
data_rows=[d for d in data_rdr] #list generator which writes DictReader object into a list to reuse data
header_rows=[h for h in header_rdr]
print(data_rows[:5])
print(header_rows[:5])

#Match up headers in data_row dict and header_row dict
for data_dict in data_rows: #iterates over each item in data_rows
    for dkey, dval in data_dict.items():
        for header_dict in header_rows: #iterates over each key & value in each data row to replace keys with more readble header labels
            for hkey, hval in header_dict.items(): #iterates over all header rows to egt readable labels
                if dkey==hval: #prints matches found between data list keys & header dictional data
                    print('match!')

#Match keys in data list with values for header row
new_rows=[] #new list to populate with clean rows
for data_dict in data_rows: #creates new dict for each row
    new_row={}
    for dkey, dval in data_dict.items():
        for header_dict in header_rows:
            if dkey in header_dict.values(): #uses values methods instead of iterating over each key:value in header rows to return list of values in dict
                new_row[header_dict.get('Label')]=dval #adds to new_row dict when there is a match
new_rows.append(new_row) #appends cleaned dict to new array
print(new_rows[:5])
```
