I used the Guardian Jobs site as my source. 

```python
# Install required packages
%pip install Requests
%pip install bs4
# Import required libraries
import requests
from bs4 import BeautifulSoup

# Retrieve HTML from web page
url="https://jobs.theguardian.com/searchjobs/?Keywords=Data+Scientist"
DataScientist=requests.get(url)
# OPTIONAL: print(page.text)

# Parse HTML content
soup=BeautifulSoup(DataScientist.content,"html.parser")
results=soup.find(id="listing")
print(results.prettify())

# Find details from job listings
job_list=results.find_all("div",class_="lister__details")
for job in job_list:
    title=job.find("h3",class_="lister__header")
    company=job.find("li",class_="lister__meta-item lister__meta-item--recruiter")
    location=job.find("li",class_="lister__meta-item lister__meta-item--location")
    print(title.text.strip())
    print(company.text.strip())
    print(location.text.strip())
    print()
    
Data_jobs=results.find_all(
    "h3",string=lambda text:"Data" in text.lower()
)
print(Data_jobs)
```
