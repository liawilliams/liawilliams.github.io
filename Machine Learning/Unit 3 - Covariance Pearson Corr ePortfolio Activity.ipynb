```python

# calculate the Pearson's correlation between two variables
from numpy import mean
from numpy import std
from numpy import cov
from numpy.random import randn
from numpy.random import seed
from matplotlib import pyplot as plt
import seaborn as sns

from scipy.stats import pearsonr
# seed random number generator
seed(1)

# prepare data
data1 = 20 * randn(1000) + 100
data2 = data1 + (10 * randn(1000) + 50)

# calculate covariance matrix
covariance = cov(data1, data2)

# calculate Pearson's correlation
corr, _ = pearsonr(data1, data2)

# plot
plt.scatter(data1, data2)
plt.show()

# summarize
print('data1: mean=%.3f stdv=%.3f' % (mean(data1), std(data1)))
print('data2: mean=%.3f stdv=%.3f' % (mean(data2), std(data2)))
print('Covariance: %.3f' % covariance[0][1])
print('Pearsons correlation: %.3f' % corr)
```

## Experiment: Changing the Data Points

The three cells above already vary the data (different spread/noise levels via the random-number multipliers). Below, one more variation is added using the same code but with data that has almost no relationship between the two variables, to see what happens to covariance and correlation when the variables are essentially independent.

```python

# calculate the Pearson's correlation between two variables
from numpy import mean
from numpy import std
from numpy import cov
from numpy.random import randn
from numpy.random import seed
from matplotlib import pyplot as plt
import seaborn as sns

from scipy.stats import pearsonr
# seed random number generator
seed(1)

# prepare data: data8 is generated completely independently of data7 (no shared term)
data7 = 20 * randn(1000) + 100
data8 = 20 * randn(1000) + 100

# calculate covariance matrix
covariance = cov(data7, data8)

# calculate Pearson's correlation
corr, _ = pearsonr(data7, data8)

# plot
plt.scatter(data7, data8)
plt.show()

# summarize
print('data7: mean=%.3f stdv=%.3f' % (mean(data7), std(data7)))
print('data8: mean=%.3f stdv=%.3f' % (mean(data8), std(data8)))
print('Covariance: %.3f' % covariance[0][1])
print('Pearsons correlation: %.3f' % corr)
```

### Summary of Observations: How Data Changes Affect Covariance & Correlation

Comparing the four variations run above:

- **data1/data2** (spread 20 for data1, plus noise term 10 for data2): covariance = 389.755, Pearson correlation = 0.888 — a strong positive relationship, since data2 was built directly from data1 with a relatively small amount of extra random noise added.
- **data3/data4** (spread widened to 40 for data3): covariance jumped to 1550.175 — much larger than data1/data2 simply because the variables now have a bigger spread, even though the correlation actually rose slightly to 0.968, since the noise term is still small relative to the widened signal.
- **data5/data6** (spread narrowed to 3 for the base signal, noise term widened to 15): both covariance (9.665) and correlation (0.208) dropped sharply compared to the first two examples, because the added random noise (15) is now large relative to the underlying signal (3), drowning out the real relationship.
- **data7/data8** (fully independent random variables, no shared term at all): covariance (8.843) and correlation (0.022) both landed very close to 0, confirming that with no real relationship between the two variables, any nonzero value is just sampling noise from the random number generator.

**Key takeaways:**
- *Covariance* is sensitive to the scale/spread of the variables — data3/data4 had by far the largest covariance simply because the values themselves were larger, not necessarily because the relationship was any stronger than data1/data2. This makes raw covariance hard to compare across data sets with different scales.
- *Pearson correlation* is a normalized version of covariance (scaled to fall between -1 and 1), so it reliably reflects the strength and direction of the relationship regardless of the variables' scale — which is why data3/data4 and data1/data2 have wildly different covariances but similarly strong correlations.
- Adding more random noise relative to the real underlying signal (data5/data6), or making the two variables fully independent (data7/data8), both push the correlation toward 0, while keeping one variable strongly derived from the other keeps the correlation high.
